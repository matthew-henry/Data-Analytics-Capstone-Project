```python
import pandas as pd
import json as json
import ujson as ujson
df = pd.read_csv('2022 AQ Clean.csv')
```

```python
#df_trans = pd.concat([df_trans, df_trans.cleaneddate.apply(ujson.loads).apply(pd.Series)], axis = 1)
expanded_datascope = df['cleaneddate'].apply(ujson.loads)
expanded_datascope = pd.json_normalize(expanded_datascope)
df = pd.concat([df, expanded_datascope], axis =1)
```

This transformation has the notebook sitting at 44GB of RAM usage.  I am going to drop a  redundant column to reduce table size, dump to disk, and continue cleaning in a separate notebook.
